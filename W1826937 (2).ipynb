# Part A - Application Area Review.

The use of Artificial Intelligence (AI) has been on the rise in different industries, and the financial markets are definately one of them. Trading has evolved over the last decade and many investors today, leverage the ability to analyse large volumes of data to make predictions about future price movement, with the use of AI subfields such as machine learning (ML), Reinforcement Learning (RL),and Natural Language Processing (NLP).[6]

Algorithmic trading refers to a trading practice used by investors by utilising computer algorithms to automate the action of selling and buying stocks, cryptocurrencies, and other financial instruments in global markets.These algorithms execute trades based on conditions that are predefined and include but aren't limited to :

    - Price fluctuations
    - Volume
    - Media News such as ETF approval deadlines etc

Algo-trading relies on historical data, statistical analysis, and mathematical models in order to identify patterns and trends, as well as potential market opportunities. It originated since the 1970s but has evolved substantially in accordance to technological advancements and computational power.

By collecting vast ammounts of financial data from various sources like historical price data or market data feeds, algo-traders use this data to form the basis for building and testing trading strategies.[4]

- risk Management

Apart from profit opportunities, algorithmic trading renders markets more liquid and trading more systematic by ruling out the impact of human emotions and human error.[5] Algorithmic trading systems also incorporate techniques to control potential losses such as stop loss orders, where a position will autmatically be closed when the price drops below a pre-defined level.

- Automated Trading
Algo-trading systems can execute trades at a much faster rate than a human can, allowing the system to take advantage of short-term opportunities which in turn leads to quicker and more consistent execution of trades.

- Improved Accuracy
AI algorithms can identify patterns that are difficult for humans to detect, which improves the accuracy of predictions.[7]

# Part B - Compare and Evaluate AI techniques

1. Sentiment Analysis - 

This type of analysis uses natural language processing (NLP) techniques to pull data from different media streams including, news articles, social media posts, and other forms of text data in order to gain insight on the current market sentiment and attitude towards a particular asset or market class.[6]

The sentiment analysis algorithm is carried out by leveraging NLP to categorise the sentiment of people about the stock value of a company into three categories such as positive, negative, and neutral. NLP is a subfield of machine learning and enables the computers to comprehend and analyse human language.[8]

    Pros -

         . Machine Learning models can be adapted to specific domains and industries making them highly versatile
           
    Cons - 
    
         . NLP models may struggle with understanding context and sarcasm leading to misinterpretation
         . The quality of sentiment analysis highly depends on the quality and bias of the training data

2. Momentum Trading Strategy -

Momentum trading is a popular algorithmic strategy which aims to capture trends in stock prices. In the context of AI techniques, momentum trading uses machine learning to build predictive models based on historical data. This requires input features such as price history, trading volumes, and various technical indicators to identify favourable entry/exit points. For example, you may choose to compare the relationship between the monthly and the weekly moving averages.[5]

This technique uses supervised learning algorithms such as linear regression, which predicts a continuous variable with the help of labelled data to train the model.

    Pros - 
    
         . Can be backtested using available historical and real-time data
         . Low latency trade order placement makes for instantaneously timed trades
         . Trades are often executed at the best possible price
   
    Cons -
    
         . Unforseen market disruptions known as black swan events, can result in unforseen losses
         . high capital costs, traders may need to pay ongoing fees for software and data feeds

 3. Neural Networks -
 
 Neural networks are a subset of machine learning and have gained popularity within algorithmic trading due to their ability to model complex relationships in data. Neural networks can automatically extract relevant features from raw data, allowing traders to focus on the most influential factors without explicitly defining them. This is an example of unsupervised learning as no predefined labels are necessary for the model to distinguish a relationship or make a prediction based on unlabelled data that it is presented with. This is particularly useful for identifying non-linear patterns.
 
     Pros - 
     
         . Neural networks can adapt to changing market conditions, making them suitable for evolving markets
         . Automatically extract relevant features from raw data, reducing the need for manual feature engineering.
    
     Cons -

         . Implementing and maintaining neural networks can involve significant costs

```python
#import libraries and framework sdk
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

#variable set to 'spdr' for our ETF
spdr = yf.download('SPY')
spdr.head()
```

Default download with no date parameters, will display data consisting of daily price action, just after the 'SPY' ETF was approved in 1993. We can customize the data range to focus the data over a specific period in time as shown later.

```python
#added a collumn labelled 'day' to help keep track what day in the timeframe we are currently at
#Use of NumPy to make a range array that matches the number of observations in our SPY dataframe
day = np.arange(1, len(spdr)+1)
spdr['day'] = day
spdr.head()
```

```python
#we can run .info on the dataframe to pull observations and display them
spdr.info()
```

# 2. Transforming/Adding data
in order to implement the momentum strategy we need to look for a signal. I will be using a moving average crossover to generate this signal. A moving average (MA),is a frequently employed stock indicator within technical analysis. Its purpose is to enhance the clarity of price data by continuously calculating an average price. An ascending moving average suggests an upward trend in the security, whereas a descending moving average signifies a downtrend[1]. We will be using this to aid us when entering a trade.

- We can either open a 'Long' position if we think the price will increase over time
- Or we can open a 'short' position if we think the price will decrease

We will be using two types of moving averages:

- Slow Moving Average
- Fast Moving Average

Simply put, a fast moving average is calculated on a shorter time-frame, therefore more reactive to change as opposed to the slow moving average, which focuses on a much longer time-frame, and tends to smooth out imbalances and price noise which is often reflected in the shorter time-frame averages.[2]

Typically we would interpret these moving averages by entering a long position when the fast moving average crosses above the slow moving average, and enter a short position when the fast moving average crosses below the slow moving average. This particular approach is especially profitable, as not only can you enter a trade when the price is predicted to go up, but also when price is predicted to go down. This way we can be prepared for either outcome by considering both the bull and bear case scenario, unlike certain indicators that only focus on uptrends.

- Add moving averages to the data frame

```python
#column added for our fast moving average at a 5 day interval (mon-fri)
spdr['5-day'] = spdr['Close'].rolling(5).mean().shift()
#column added for our slow moving average at a 21 day interval (trading days in a month)
spdr['21-day'] = spdr['Close'].rolling(21).mean().shift()
#trading indicators are flawed in the sense that they lag behind real-time data. In order to calculate the 21 day average based on the closing price that day, we have to use the .shift command to delay our calculation by 1 day.
spdr[19:25]
```

- Add 'signal' column to tell us when we would enter a long or short trade.

```python
#np.where works like an if statement in excel
spdr['signal'] = np.where(spdr['5-day'] > spdr['21-day'], 1, 0)
#When the 5-day average crosses above the 21-day average it will print a '1' else it will print a '0'. This will be our buy signal
spdr['signal'] = np.where(spdr['5-day'] < spdr['21-day'], -1, spdr['signal'])
#when the 5-day average crosses below the 21-day average it will print a '-1' else it will leave whatever is in that signal there
spdr.dropna(inplace=True)
#this will drop off the rows which can't be used to calculate our MA and would typically print a NaN signal
spdr.head()
```

- Next we need to evaluate the performance of SPY and compare instantaneous returns in respect to our systems performance

```python
#This shows us the typical return achieved from buying and holding a security
spdr['hold'] = np.log(spdr['Close']).diff()
#this shows our system return
spdr['sys_return'] = spdr['signal'] * spdr['hold']
#added column to display value when we change trend from long--->short (-2), and from short--->long (2) positive/negative
spdr['entry'] = spdr.signal.diff()
spdr.head()
```

Notice we do not get a return for the first day because the indicator needs to calculate the change between the first day and the next

# 3. Visualise by plotting trades on time series
 we plot a line plot for each of those 3 colums (closing price, fast moving average, slow moving average)

```python
plt.rcParams['figure.figsize'] = 14,10
plt.grid(True, alpha = .2)
plt.plot(spdr.iloc[-252:]['Close'], label = 'SPY')
plt.plot(spdr.iloc[-252:]['5-day'], label = '5-day')
plt.plot(spdr.iloc[-252:]['21-day'], label = '21-day')
#plots upward facing arrowhead thats green when the entry difference is equal to 2
plt.plot(spdr[-252:].loc[spdr.entry == 2].index, spdr[-252:]['5-day'][spdr.entry == 2], '^', 
        color = 'g', markersize = 5)
#plots downward facing arrowhead thats red when the entry difference is equal to -2
plt.plot(spdr[-252:].loc[spdr.entry == 2].index, spdr[-252:]['21-day'][spdr.entry == 2], 'v', 
        color = 'r', markersize = 5)
plt.legend(loc = 2)
```

# 4. optimisation
Our linear chart is plotted over a period of 252 days which is the amount of trading days in a year. The data presented dates back a year until present (2023-01 - 2024-01). # 5. Testing
now we want to evaluate the performance of our signal indicator,against a more traditional approach to investing such as the buy and hold method.

```python
spdr_last_5_years = spdr.last('5Y')

plt.plot(np.exp(spdr_last_5_years['hold']).cumprod(), label='buy/hold')
plt.plot(np.exp(spdr_last_5_years['sys_return']).cumprod(), label='system')
plt.legend(loc=2)
plt.grid(True, alpha=0.3)
plt.show()
```

# 6. Evaluation
As we can see, over the last 4 years, the indicator has mostly outperformed an investor who would use a common buy and hold strategy when investing in the SPY security. Our implementations performance is largely credited due to being optimized and having the capability of making highly probable trades even when the market sentiment is low and trend is declining by opening short positions as well as long trades, which a typical investor would not have the resources for. Even a more experienced trader who trades futures, could easily get liquidated, however the indicator helps to avoid this by signaling when the trend is changing on both short and longer time frames using the moving average indicator.

However, we need to also take into account that most if not all market indicators are flawed due to them 'lagging' behind, as they need to build up sufficient historical data before they print a signal, making them inconsistent. This means they also tend to perform worse over shorter time frames due to the nature of volatility that occurs there. It is advised that such indicators should be used as a confluence to other market analysis techniques and not solely as an individual entity.

To get around this, we can pull real-time data from price feeds that are updated at a more frequent rate in order to expand our dataset and train the model further, however at the expense of higher load on system resources such as processing power and memory usage.
